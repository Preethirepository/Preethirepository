```python
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get("https://en.wikipedia.org/wiki/Comma-separated_values")
```

```python
soup = BeautifulSoup(req.content, "html.parser")

print(soup.get_text())
```

```python
res = soup.title
print(res.prettify())
```
